```python
# Importing the required libraries
# JSON will read the JSON files containing the streaming history

# Importing the JSON library so the .json file can be read
import json

# Pandas is needed to convert the data into a dataframe in order to save to CSV 
import pandas as pd

# Here im defining the relative path of the JSON file that contains the streaming history for 2023
# I proceed to open the JSON file in 'r' read mode and load it into a python dictionary in key:value pairs
streaming_history_file_path = 'StreamingHistory_music_0.json'
with open(streaming_history_file_path, 'r') as file:
    streaming_history_data = json.load(file)

# Converting the data from the JSON file containing the streaming history to a pandas dataframe
streaming_history_df = pd.json_normalize(streaming_history_data)

# Here I am defining the pathname for the CSV file that our data will be contained in
# Once initialised the data from the JSON file has been saved to a CSV file
streaming_history_csv_file_path = 'streaming_history_music_0_data.csv'
streaming_history_df.to_csv(streaming_history_csv_file_path, index=False)

# Printing out the path of the new CSV file
# this was added to make sure that the code was working as intended
streaming_history_csv_file_path

# In these few lines of code I'm utilising the milliseconds provided in order to get the minutes and hours

# Converting the miliseconds to seconds by dividing by 1000
streaming_history_df['secondsPlayed'] = streaming_history_df['msPlayed'] / 1000
# Converting seconds to minutes by dividing seconds played by 60
streaming_history_df['minutesPlayed'] = streaming_history_df['secondsPlayed'] / 60
# Converting minutes to hours by dividing the minutes played by 60
streaming_history_df['hoursPlayed'] = streaming_history_df['minutesPlayed'] / 60

# Using the head fucntion in order to display the first few rows of the data
# This is to make sure that the conversions are correct and code worked as intended.
streaming_history_df.head()

# Checking for duplicates
duplicates = streaming_history_df[streaming_history_df.duplicated()]

# Counting the number of duplicate rows / repeated songs
duplicate_count = duplicates.shape[0]

duplicate_count

# Intialising the path for the streaming history CSV file and writing the data we extracted to the CSV file
updated_streaming_history_csv_file_path = "streaming_history_music_0_data_updated.csv"
streaming_history_df.to_csv(updated_streaming_history_csv_file_path, index=False)

updated_streaming_history_csv_file_path


# Group by 'artistName' and 'trackName', and sum the 'secondsPlayed', 'minutesPlayed', and 'hoursPlayed'
grouped_streaming_history = streaming_history_df.groupby(['artistName', 'trackName']).agg({
    'secondsPlayed': 'sum',
    'minutesPlayed': 'sum',
    'hoursPlayed': 'sum'
}).reset_index()

# Using the .head  function in order to verify that the aggregation code above worked so artist were grouped 
# Along with the duration of time spent listneing 
grouped_streaming_history.head

# Intialising the path name for the CSV file that contains the grouped listening data
# Saving the grouped / aggregated Spotify data to a CSV file
grouped_streaming_history_csv_file_path = "grouped_streaming_history.csv"
grouped_streaming_history.to_csv(grouped_streaming_history_csv_file_path, index=False)

# printing again the name of the CSV file, this was just to make sure that the code is working as intended.
grouped_streaming_history_csv_file_path
```

## Getting a more accurate location for each artist

## Code Breakdown

<b> What this code is doing </b>
1. Reading in a txt file containing the artist names and their location (city, state, country)
2. I then parse the artist name(s) along with their locations seperating into city, state and country
3. I proceed to then merge the parsed location data with the existing artist data based on the name of the artist, along with adding the city, state and country information to the dataframe
4. Once merged the data is saved to a CSV file to be used in the next cell that utilises the geopy library to get the coordinates (latitude & longitude)

```python
# Importing the pandas library so the data can be extracted, manipulated and saved to another CSV
import pandas as pd


# Here I am defining the file path locations for all of the CSV file's we will be using
# Definining: The txt file with artist locations, CSV that contains all the unique artists and CSV of the complete info
artist_city_location_text_file_path = 'Artist City and Location.txt'
artist_locations_csv_file_path = 'unique_artists.csv'
merged_csv_file_path = 'merged_artist_locations.csv'

# Initialising a list that will contain all of the parsed data
parsed_data = []

# Performng reading and parsing on the artist location text file
with open(artist_city_location_text_file_path, 'r') as file:
    # looping through each line within the text file
    for line in file:
        # Split each line by " - " to separate artist from location
        artist, location = line.strip().split(" - ")
        # Split the location by ", " to separate city, state, and country
        location_parts = location.split(", ")
        # Extracting the city, state and location from the txt file
        city = location_parts[0] if len(location_parts) > 0 else 'Unknown'
        state = location_parts[1] if len(location_parts) > 1 else 'Unknown'
        country = location_parts[2] if len(location_parts) > 2 else 'Unknown'
        
        # Add the parsed data to the list
        parsed_data.append({'artistName': artist, 'City': city, 'State': state, 'Country': country})

# Using pandas to convert the parsed data into a data frame
parsed_df = pd.DataFrame(parsed_data)

# Using pandas to read the artist locations CSV file
unique_artists_df = pd.read_csv(artist_locations_csv_file_path)

# Performing a merge of both the parsed data with the existing data using the 'artistName' column as the key value
merged_df = pd.merge(unique_artists_df, parsed_df, on='artistName', how='left')

# Saving the merged data to the new CSV file which will contain all the information related to the artist and location
merged_df.to_csv(merged_csv_file_path, index=False)

# I added a print statement that was a success message that the merged CSV file was created
# Primairly used as a print statement to know that the python worked as expected
print(f'Merged artist location data has been saved to {merged_csv_file_path}')

```

### Longitude and Latitiude for each Artist

## Breakdown of code

<b>Geopy Library</b>

In order to get the latitude and longitude coordinates I utilised the geopy library which allows for geocoding and reverse geocoding. In order for this library to function correctly it utlises the city, state and location variables within the dataset.

<b>What is this code doing?</b>

1. This code utilises the pandas, time, geopy libraries to allow for geocoding
2. The artist dataset is read in using the pandas library to a dataframe
3. I then proceed to create an instance of the Nominatim geocoder with my own user agent 'EdwardsGeocoder' and set for a timeout so I don't send too many pings to this server.
4. I applied a rate limiter to prevent to avoid overloading the API and be respectful of other users.
5. I then used the 'apply_geocode' function to construct the location string from the city, state and country variables for each of the rows and apply the geocoding function to get the coordinates (latitiude and longitude) for each instance.
6. I then proceed to add the longitude and latitiude to the pandas dataframe and save it to a new CSV file which will now contain all of the necessary variables to create map visualisations of the dataset.
7. Just to make sure that the code ran successfully I added a short print statement at the end that told us that the CSV file was created with the addition of coordinates.


```python
# Importing the libraries we need

# Pandas will be used to manipulate the dataset and create a dataframe to update the new CSV
import pandas as pd

# Importing the time module for managing time
import time 

# Imprting the Nominatim geocoder as shown in the practicals 
from geopy.geocoders import Nominatim

# Importing the rate limiter so I'm not inadvertnly overloading the API with requests
from geopy.extra.rate_limiter import RateLimiter

# Imprting the GeocoderUnavailable excpetion module to handle instances where coordinates can't be obtained
from geopy.exc import GeocoderUnavailable  

# Using pandas I'm reading in the dataset containing the artist locaitons
cleaned_artist_dataset_relative_path = 'cleaned_artist_data.csv'
df = pd.read_csv(cleaned_artist_dataset_relative_path)

# Initializing the geocoder 'Nominatim' with the user-agent 'EdwardsGeocoder' 
# with a timeout of 10 seconds to prevent overloading the API
geolocator = Nominatim(user_agent='EdwardsGeocoder', timeout=10)

# Applying a rate limiter to avoid overloading the geocode API
# Setting a minimum delay of 2 seconds between each geocoding request
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=2)

# Creating a function that will apply geocoding / adding latitude and longitude
# Using the city, state and country variables
def get_location_by_geocode(address):
    try:
        # Geocoding the provided address using the rate-limited geocoder
        location = geocode(address)
        # If the location is found by geopy, the latitide and longitude is returned
        if location:
            return (location.latitude, location.longitude)
        return (None, None)  # Return None if location is not found
    except GeocoderUnavailable as e:
        # Here I'm handling cases where if the coordinates can't be found by printing an error message, giving it 10 seconds and trying again.
        print(f"Geocoding error: {e}, retrying...")
        time.sleep(10)
        return geocode(address)


# Creating a function 'apply_geocoding' that will apply the new coordinates to the orignal dataframe
def apply_geocoding(row):
    # # Constructing the location string using city, state, and country information from each row of the DataFrame
    location_str = f"{row['City']}, {row['State']}, {row['Country']}"
    # Calling the get_location_by_geocode function to perform geocoding and returning latitude and longitude
    return get_location_by_geocode(location_str)

# Apply the function to each row in the DataFrame and expand the results into separate columns
df[['latitude', 'longitude']] = df.apply(apply_geocoding, axis=1, result_type='expand')

# Applying the apply_geocoding function to each row in the DataFrame
# Expanding the results into separate latitude and longitude columns in the DataFrame
# Created the file path for the new CSV file containing all of the location details of each artists
updated_csv_file_path = 'Updated_artist_data_with_coordinates.csv'
df.to_csv(updated_csv_file_path, index=False)

# Printing a success message that the CSV file was created, coordinate retreival was successful
# code was ran without any issues
print(f"Updated artist data with coordinates has been saved to {updated_csv_file_path}")
```

## Creating a new dataset from the JSON file provided by Spotify

## Breakdown of the code

<b> What is this code doing? </b>

1. Import the required libraries: pandas for data manipulation and calendar for extracting month names.
2.I then rRea thed streaming history data from multiple JSON files into separate pandas DataFrames.
3. Combingne the individual DataFrames into a single DataFrame.
I proceed to 4. Convert the 'endTime' column to datetime format for easier manipulation.
5. Extingr the act date and time components from the 'endTime' column.I then e
6. Extract additioral features such as the hour, day of the week, month, month name, and from each of the columns of the dataset year.
7. Convert the 'msPlayed' column to seconds and minutes for easier an so we can create better plots and easier interpretabilityalysiI then s.
8all of . Save the processed streaming history data to a C called 'processed_streaming_history.csv'SV file. .

```python
import pandas as pd
import calendar

# Reading in both the JSON files containing your streaming history using pandas
json_file_paths = ['StreamingHistory_music_0.json','StreamingHistory_music_1.json']
data_frames = []

# I proceed to iterate through each of the JSON files
# 1. Open each JSON file
# 2. Read the JSON files into a single pandas dataframe
# 3. Add the dataframe to the data_frames list variable
for file_path in json_file_paths:
    with open(file_path,'r',encoding='utf-8') as file:
        data = pd.read_json(file)
        data_frames.append(data)

# Combining both of the streaming history dataframes into a single DataFrame 
# Adding all the dataframes into a single dataframe using the concat function
# I added a print message so we know that the operation was successful 
streaming_history_dataframe = pd.concat(data_frames, ignore_index=True)
print('Streaming history DataFrame has been created')

# Converting the endTime to datetime
streaming_history_dataframe['endTime'] = pd.to_datetime(streaming_history_dataframe['endTime'])

# Extracting the date and time components of the data from each of the columns
# 1. I extract the date the song was played
# 2. I extract the time the song was played
# 3. I extract the hour the song was played
# 4. I extract the day of the week the song was played
# 5. I extract the month the song was played
# 6. I extract the year the song was played
streaming_history_dataframe['date'] = streaming_history_dataframe['endTime'].dt.date
streaming_history_dataframe['time'] = streaming_history_dataframe['endTime'].dt.time
streaming_history_dataframe['hour'] = streaming_history_dataframe['endTime'].dt.hour
streaming_history_dataframe['day_of_week'] = streaming_history_dataframe['endTime'].dt.day_name()
streaming_history_dataframe['month'] = streaming_history_dataframe['endTime'].dt.month  # Extracting the month
streaming_history_dataframe['month_name'] = streaming_history_dataframe['month'].apply(lambda x: calendar.month_name[x])
streaming_history_dataframe['year'] = streaming_history_dataframe['endTime'].dt.year

# Convert msPlayed to seconds and minutes for easier analysis
# 1. I converted the milliseconds to seconds
# 2. Converted the seconds to minutes
streaming_history_dataframe['secondsPlayed'] = streaming_history_dataframe['msPlayed'] / 1000
streaming_history_dataframe['minutesPlayed'] = streaming_history_dataframe['secondsPlayed'] / 60

# Creating a CSV file containing our data
# 1. I defined the file path for this new dataset
# 2. I then saved this dataframe containing all of these new variables to a CSV file
streaming_history_csv_file_path = 'processed_streaming_history.csv'
streaming_history_dataframe.to_csv(streaming_history_csv_file_path, index=False)

print(f'Processed streaming history data has been saved to {streaming_history_csv_file_path}')
```

```python
import requests
import pandas as pd

#Fetch audio features for a given Spotify ID.
#Args:
#spotify_id (str): The Spotify ID of the track.
#access_token (str): The access token for authentication.
#Returns:
# dict: A dictionary containing audio features of the track.
def get_audio_features(spotify_id, access_token):
    """Fetch audio features for a given Spotify ID."""
    url = f"https://api.spotify.com/v1/audio-features/{spotify_id}"
    headers = {'Authorization': f'Bearer {access_token}'}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    elif response.status_code == 429:  # Handling rate limit specifically
        retry_after = int(response.headers.get('Retry-After', 20))  # Defaulting to 20 seconds if header is missing
        print(f"Rate limit exceeded, waiting for {retry_after} seconds.")
        time.sleep(retry_after)
        return get_audio_features(spotify_id, access_token)  # Recursively retry fetching audio features
    else:
        print(f"Error fetching audio features for {spotify_id}: {response.status_code}")
        return {}

# Loaded in my dataset of tracks with Spotify ID's
df = pd.read_csv('updated_streaming_history_with_ids.csv')

# created a dictionary to store audio features for known Spotify IDs, this prevented repeated API calls and decreased usage
audio_features_lookup = {}

# Prepare to store audio features in the DataFrame if not already present
audio_features_columns = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
for column in audio_features_columns:
    if column not in df.columns:
        df[column] = None

# Using the access token I got from the previous cell block
access_token = 'BQDK_HkEsLmuZOGT8Kuq2VzB3U2cZDDGMsDr3qQW3T9wZVoFYmcEMaH0wojEWsSNGRrxJfvcuadL7x58yC2UC9NpS3xKfrQnjTBW1WlAYvS4iH_TmKI'

# I Defined the batch size and start from the first unprocessed track, this was to save the progress as processing 14,000 tracks at once is impossible 
batch_size = 1000
total_tracks = len(df)
for start in range(7000, total_tracks, batch_size):
    end = min(start + batch_size, total_tracks)
    print(f"Processing tracks {start+1} to {end}")

    for index, row in df.iloc[start:end].iterrows():
        if pd.notnull(row['spotify_id']):
            if row['spotify_id'] not in audio_features_lookup:
                audio_features = get_audio_features(row['spotify_id'], access_token)
                if audio_features:
                    audio_features_lookup[row['spotify_id']] = audio_features
                    for feature in audio_features_columns:
                        df.at[index, feature] = audio_features.get(feature)
                    print(f"Fetched and cached new audio features for {row['trackName']} by {row['artistName']}")
            else:
                # Use cached audio features
                for feature in audio_features_columns:
                    df.at[index, feature] = audio_features_lookup[row['spotify_id']][feature]
                print(f"Using cached audio features for {row['trackName']} by {row['artistName']}")

    # Save progress after each batch
    save_path = f'streaming_history_with_audio_features_{start+1}_{end}.csv'
    df.iloc[start:end].to_csv(save_path, index=False)
    print(f"Saved audio features for tracks {start+1} to {end}")

print("Audio features have been successfully added to the DataFrame for all tracks.")
```
